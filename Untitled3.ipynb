```python
#defining the function 
def Details(data, col_1, col_4 ):
    '''
    The above function is used to create a line plot for Online Sales data
    
    Parameters:
    - data: DataFrame containing the data.
    - col_3: Name of the column for Quantity.
    - col_2: Name of the column for Profit.
    
    Returns:
    Results show a line plot between Quantity and Profit.

    '''
    
 #plotting data to show in the line graph   
    sns.lineplot(x='Quantity', y='Profit', data=ol, errorbar=('ci', False), hue='Category', marker='o')

    plt.grid()
    plt.xlabel('Quantity')
    plt.ylabel('Profit')
    plt.title('Quantity vs Profit')

    plt.show()
#calling the function
Details(ol, 'Quantity' , 'Profit')
```

```python
#defining the function
def Details(data, col_5, col_1):
    '''
    The above function is used to create a bar graph for Online Sales data
    
    Parameters:
    - data: DataFrame containing the data.
    - col_5: Name of the column for sub_category.
    - col_1: Name of the column for Amount.
    
    Returns:
    Results show a bar graph between sub_category and Amount.
    '''
#plotting the data for showing the bar graph
    plt.figure(figsize = (20,8))
    sns.barplot(x='sub_category', y='Amount', data=ol, errorbar=None)

    plt.xlabel('sub_category', labelpad=14)
    plt.ylabel('Amount', labelpad=14)
    plt.title('Total Amount by Sub Category')

    plt.show()
#calling the function
Details(ol, 'sub_category', 'Amount')
```

```python
#defining the function  
def Details(data, col_6, col_2):
    '''
    The above function is used to create a box plot for Online Sales data
    
    Returns:
    Results shows box plot between payment mode and Profit
    
    '''

    sns.boxplot(x = 'payment_mode', y = 'Profit', data = ol, showfliers = False)

#plotting data to create a box plot    
    plt.xlabel('Payment Mode', labelpad = 14)
    plt.ylabel('Profit', labelpad = 14)
    plt.title('Distribution of Profit by Payment Mode')

    plt.tight_layout()
    plt.show()
#calling the function    
Details(ol, 'payment_mode', 'Profit')
```
